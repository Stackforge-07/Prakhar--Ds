```python
import numpy as np
print("Enter total saman u want to buy")
total=int(input())
saman=[]
for i in range(0,total):
    saman_name=str(input())
    saman.append(saman_name)

print(saman)
saman_price=[]
for i in range(0,total):
    x=int(input())
    saman_price.append(x)

print(saman_price)

lst=np.array(saman)
price=np.array(saman_price)
 
for i in range(0,total):
    print(f"Saman_name is {lst[i]} ,"" ""Price of saman is :",price[i])


mini=np.min(price)
print("Min price saman is: ",mini)

maxi=np.max(price)
print("Max price saman is :",maxi)

sumi=np.sum(price)
print("Total daam is :", sumi ,"+" " GST bhi lagega ruko avi")
```
