```python
str="Prakhar Raj"
str1="   New    "
print(len(str))
print(str.upper())
print(str.lower())
print(str.capitalize()) # first letter capital kar dega
print(str.title()) # har word ka pehla letter capital kar dega
print(str.split())
print(str1.strip()) # ye aage and peeche se space hatata hai b

a=" today "
b="  weather "
c="is Good "
a=a.lstrip()
b=b.lstrip()
c=c.rstrip().lower()

print(a)
print(b)
print(c)
d=a+b+c
print(d)

d=d.replace("good","Bad").lower()
print(d)

# Find index me difference ni hai bas find to ko ni mila to yo -1 dega or index error dega
print(d.find("bad"))
print(d.index("ad")) 

#starts with ends with true ya false dega batane ke liye humre string start ya end particular word se ho raha ya ni
print(d.startswith("t"))

print(d.endswith("t"))

print(d.count("a"))
print(sorted(d))
```
