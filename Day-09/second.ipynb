```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('./student_data.csv',sep=',')

def grade(a):
    if(a>=90):
        return 'A'
    elif(a>=70):
        return 'B'
    elif(a>=50):
        return 'C'
    elif(a>=30):
        return 'D'
    else:
        return 'F'

df['grade']=df['exam_score'].apply(grade)
display(df)

```

```python
sns.lineplot(data=df,x='study_hours',y='exam_score')
plt.tight_layout()
plt.show()
#shadow ye bata raha hai ki data is range me aa raha hai
#matplot se bhi kar sakte hai but isme complex kam hai and easy hai
```

```python
sns.barplot(df,x='grade',y='exam_score',estimator='mean') # estimator batayega category me kya karna hai 
plt.title("Performance")
plt.tight_layout()
plt.show()
#black line precision bata raha hai
```
