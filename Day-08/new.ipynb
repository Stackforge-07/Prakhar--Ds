```python
class person:

    def __init__(self,h,w,age,col):
        self.height=h
        self.weight=w
        self.age=age
        self.color=col



class Gautam(person):

    def __init__(self,g,n,b,h,w,age,col):

        super().__init__(h,w,age,col)
        self.gender=g
        self.name=n
        self.bankbalance=b 
        #self ni lagaya to access ni hoga
    

gtm=Gautam('Not KNown','Gautam','$20',100,30,20,'Brownie')

print('Gender is:',gtm.gender)
print('Name is :',gtm.name)
print('BB is :',gtm.bankbalance)
print('height is :',gtm.height)
print('weight is :',gtm.weight)
print('age is :',gtm.age)
print('coloe is :',gtm.color)




        


```
