```python
import pandas as pd
df=pd.DataFrame({
    "id":[1,2,3,4],
  "  name":["A","b","c","d"],
  "age":[20,10,20,15]

})


df1=pd.DataFrame({
    "id":[3,4,5,6],
  "name":['e','f','g','h'],
  "age":[18,20,19,25]

})

df3=pd.merge(df,df1,on='id',how='inner')
df3.rename(columns={'age_x': 'table_age1'},inplace=True)
# display(df3)

df3=pd.merge(df,df1,on='id',how='left')
# print(df3)

df3=pd.merge(df,df1,on='id',how='right')
# display(df3)

df3=pd.merge(df,df1,on='id',how='outer')
# display(df3)

df3=pd.merge(df,df1,how='cross') # har row har ek row ke sath cross ho raha hai to 16 relation banega o isliye ni hota bcz sabke sabke sath combn hota hai
# display(df3)

df3=pd.concat([df,df1],ignore_index=True) #apply on same number of column
# display(df3)




```
