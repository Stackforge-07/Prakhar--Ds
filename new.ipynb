```python
lst=[1,2,3,'name',[2,4]]
#append (directly last me bhej dega)

lst.append(3)
print(lst)

# lst.append(1,2,4) list.append() takes exactly one argument (3 given)

lst.append([1,4,7])
print(lst)

#Extend function (serate element jayega)
lst.extend([2,3,4])
print(lst)

#insert
lst.insert(2,"yash")
print(lst)
```

```python
#Remove Function from list
lst=[1,2,3,2,2,'name',[2,4]]
#pop removes last element
lst.pop()
print(lst)

#remove (it removes values of first occurance not index from the list)
lst.remove(2)
print(lst)



#index (it return index)
print(lst.index(1))


#count
print(lst.count(2))

#sort ,sum ,min,max funtion:
#sort return kuch ni karta
lsti=[2,3,1,2,1,13,3]
lsti.sort()
print(lsti)

#string me asci value pe sort karega
stri=["name","yash","rj"]
stri.sort()
print(stri)

lste=[1,4,344,34,34,34,3]
#sum method ni hai proper function hai
#same sorted proper fucntion hai jo proper list return kar raha hai and sort return ni karta ye bas purana list ko sort kar deta hai

print(sum(lste))
newlste=sorted(lste)
print(newlste)
```

Tupple
immutable ek bar define kar diya to kardiya 
to tupple me un fucntion ko use karenge jo return kar sakte hai
list me chnage kar sakte hai add, remove  hence mutable hote hai 

```python
tp=(1,2,2,3,4,5,6,7,8)
print(tp)

#tp.sort() it changes the order to use ni kar sakte
#jo function value alter ni karega unko tupple me use karenge

sor=sorted(tp) #--> ye list return karta hai
sort=tuple(sorted(tp)) #returns tuple
print(sor)
print(sort)
#function in tupple
#count()
#index()
#len()
#sum()
#max(),min()
#sorted()

print(tp.count(2))
print(tp.index(2))
print(len(tp))
print(sum(tp))
print(min(tp))
print(max(tp))

```
