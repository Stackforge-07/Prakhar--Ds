```python
import numpy as np
array=np.random.randint(1,91,size=(5,5))
print(array)



rmean=np.mean(array,axis=1)
print(rmean)

cmean=np.mean(array,axis=0)
print(cmean)



replace=np.where(array%2==0,-1,array)
print(replace)


csrt=np.sort(array,axis=0)
print(csrt)

rsrt=np.sort(array,axis=1)
print(rsrt)


rsum=np.sum(array,axis=1)
print(rsum)


csum=np.sum(array,axis=0)
print(csum)



cnt_col=np.sum(np.where(replace<0,1,0),axis=0)
print(cnt_col)
```

```python
import numpy as np
stock_level=np.random.randint(1,91,size=(5,5))

demand=np.random.randint(1,91,size=(5,5))


remaning_stock=stock_level-demand
print(remaning_stock)
print("\n")


out_stock=np.where(remaning_stock<0,0,remaning_stock)
print(out_stock)
print("\n")



zeroshortage=np.sum(np.where(out_stock==0,1,0),axis=1)
print(zeroshortage)
print("\n")


totalshortage=np.sum(np.where(remaning_stock<0,remaning_stock,0),axis=1)
print(totalshortage)
print("\n")


rsort=np.sort(remaning_stock,axis=1)
print(rsort)




```
