```python
df.groupby('study_hours')['exam_score'].agg('mean')

def grade(a):
    if(a>=90):
        return 'A'
    elif(a>=70):
        return 'B'
    elif(a>=50):
        return 'C'
    elif(a>=30):
        return 'D'
    else:
        return 'F'

df['grade']=df['exam_score'].apply(grade)
display(df)
    
```
